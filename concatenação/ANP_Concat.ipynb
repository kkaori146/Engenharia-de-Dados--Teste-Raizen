#PySpark

```python
# Construção do Sparksession
spark = ( 
    SparkSession.builder
                .master("local")
                .appName("dataset_derivados_pandas")
                .config("spark.ui.port", "4050")
                .getOrCreate() 
        )
```

```python
# StructType
esquema = (
    StructType([
        
        StructField('ano_mês', DateType()),
        StructField('uf', StringType()),
        StructField('produto', StringType()),
        StructField('unid', StringType()),
        StructField('venda', DoubleType()),
        StructField('criado_em', TimestampType())
                
    ]))
```

```python
# Ao final do arquivo csv, foi adicionado o schema do dataframe construído no StructType

dfddp = ( spark.read.format("csv")
                  .option("header", "true")
                  .option("inferschema", "true")
                  .option("delimiter", ",")
                  .load('/content/dataset_derivados_pandas.csv', schema = esquema)
)
```

```python
# Análise dos tipos das colunas

dfddp.printSchema()
```

```python
# Criando a lista de Colunas Novas

colunasNovas = []

for i in dfddp.columns:
  colunasNovas.append(i)
  print(i)
```

```python
# Verificando inconsistencias por valores unicos no topo e final das colunas

for i in range(len(colunasNovas)):
  dfddp.select(F.col(colunasNovas[i])).distinct().orderBy(colunasNovas[i], ascending=True).show(10)
```

```python
# Verificando inconsistencias por valores unicos no topo e final das colunas

for i in range(len(colunasNovas)):
  dfddp.select(F.col(colunasNovas[i])).distinct().orderBy(colunasNovas[i], ascending=False).show(10)
```

```python
# Tratamento de inconsistencias na coluna volume

dfddp = dfddp.replace(str('nan'), None)
```

```python
dfddp = dfddp.replace(str('null'), None)
```

```python
# Eliminando todos os na

dfddp=dfddp.dropna()
```

```python
# Contagem das inconsistencias

dfddp.where(F.col('venda').isNull()).count()
```
