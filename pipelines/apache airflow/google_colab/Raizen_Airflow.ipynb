### Importação da biblioteca

```python
import pandas as pd
import datetime
```

## Derivados Petróleo


```python
# Leitura do dataset

dfpetrol = pd.read_csv('/content/vendas-derivados-petroleo-etanol-m3-1990-2022.csv', sep=";")
```

```python
# Visão geral do dataset

dfpetrol
```

```python
# Adição da coluna unit

dfpetrol['unit'] = 'm3'
```

```python
# Criação da coluna "created_at"

dfpetrol['created_at'] = datetime.datetime.now()
```

```python
# Verificação do nome das colunas

dfpetrol.head(2)
```

```python
# Troca do formato string por numérico na coluna meses

dfpetrol['MÊS'].replace([
    'JAN',
    'FEV',
    'MAR',
    'ABR',
    'MAI',
    'JUN',
    'JUL',
    'AGO',
    'SET',
    'OUT',
    'NOV',
    'DEZ'],
    ['01',
     '02',
     '03',
     '04',
     '05',
     '06',
     '07',
     '08',
     '09',
     '10',
     '11',
     '12'], inplace=True)
```

```python
# Coluna MÊS

mes =  sorted(pd.unique(dfpetrol['MÊS']))
for m in range(len(mes)):
  if m % 20 == 0:
        print()
  print(mes[m], end=', ')
```

```python
# Verificação do dataframe após modificações

dfpetrol.head(5)
```

```python
# Renomeando as colunas

dfpetrol.rename(columns={"UNIDADE DA FEDERAÇÃO": "uf", "PRODUTO": "product", "VENDAS": "volume" }, inplace=True)
```

```python
# Replace dos nomes dos estados

dfpetrol['uf'].replace([
    'ACRE',
    'ALAGOAS',
    'AMAPÁ',
    'AMAZONAS',
    'BAHIA',
    'CEARÁ',
    'DISTRITO FEDERAL',
    'ESPÍRITO SANTO',
    'GOIÁS',
    'MARANHÃO',
    'MATO GROSSO',
    'MATO GROSSO DO SUL',
    'MINAS GERAIS',
    'PARANÁ',
    'PARAÍBA',
    'PARÁ',
    'PERNAMBUCO',
    'PIAUÍ',
    'RIO DE JANEIRO',
    'RIO GRANDE DO NORTE',
    'RIO GRANDE DO SUL',
    'RONDÔNIA',
    'RORAIMA',
    'SANTA CATARINA',
    'SERGIPE',
    'SÃO PAULO',
    'TOCANTINS'],
    ['AC',
     'AL',
     'AP',
     'AM',
     'BA',
     'CE',
     'DF',
     'ES',
     'GO',
     'MA',
     'MT',
     'MS',
     'MG',
     'PR',
     'PB',
     'PA',
     'PE',
     'PI',
     'RJ',
     'RN',
     'RS',
     'RO',
     'RR',
     'SC',
     'SE',
     'SP',
     'TO'], inplace=True)
```

```python
# Coluna uf

uf =  sorted(pd.unique(dfpetrol['uf']))
for u in range(len(uf)):
  if u % 20 == 0:
        print()
  print(uf[u], end=', ')
```

```python
# Concatenando os dados das colunas ANO e Mês

dfpetrol['year_month'] = dfpetrol['ANO'].map(str) + '-' + dfpetrol['MÊS'].map(str)
```

```python
# Verificação do dataframe após modificações

dfpetrol.head(5)
```

```python
# Organização das colunas

dfpetrol = dfpetrol[[
    'year_month',
    'uf',
    'product',
    'unit',
    'volume',
    'created_at'
]]

```

```python
# Verificação do dataframe após modificações

dfpetrol.head(5)
```

```python
# Definindo atributo na coluna year_month

dfpetrol['year_month'] = pd.to_datetime(dfpetrol['year_month'])
```

```python
# Verificação do dataframe após modificações

dfpetrol.head(5)
```

```python
# Replace de vírgula por ponto

dfpetrol['volume'] = dfpetrol['volume'].str.replace(",",".")
```

```python
# Definindo os tipos de colunas

dfpetrol['volume'] = dfpetrol['volume'].astype(float)
```

```python
# Coluna volume

volume =  sorted(pd.unique(dfpetrol['volume']))
for v in range(len(volume)):
  if v % 20 == 0:
        print()
  print(volume[v], end=', ')
```

```python
# Verificando os tipos de Colunas

dfpetrol.dtypes
```

### GroupBy para obter os datasets solicitados

```python
# Volume total dos derivados de petroleo vendidos para todos os estados 

dfderivados = dfpetrol.groupby(['uf','product']) ['volume'].sum().reset_index() 
```

```python
dfderivados.head(100)
```

```python
# Volume total de ÓLEO DIESEL vendido para os estados 

dfdiesel = dfderivados[dfderivados['product'] == 'ÓLEO DIESEL']
```
