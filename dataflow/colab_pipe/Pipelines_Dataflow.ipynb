##Quarta Tentativa

```python
# Filtrando pelo produto e estado (retornando duplicidade de dados)

p1 = beam.Pipeline()

produtos = [
          'Etanol Hidratado',
          'Gasolina de Aviação',
          'Gasolina C',
          'GLP',
          'Óleo Combustível',
          'Óleo Diesel',
          'Querosene de Aviação',
          'Querosene Iluminante'
]

ufs = [
  
  'RO',
  'AC',
  'AM',
  'RR',
  'PA',
  'AP',
  'TO',
  'MA',
  'PI',
  'CE',
  'RN',
  'PB',
  'PE',
  'AL',
  'SE',
  'BA',
  'MG',
  'ES',
  'RJ',
  'SP',
  'PR',
  'SC',
  'RS',
  'MS',
  'MT',
  'GO',
  'DF'
]



for j in range(len(ufs)):

  for i in range(len(produtos)):
    
    derivados = (
        p1
        |f'Extrair do CSV{i}{j}'>> beam.io.ReadFromText('/content/dataset_final_derivados.csv',skip_header_lines = 1)
        |f'Separador do CSV{i}{j}'>> beam.Map(lambda record: record.split(','))
        |f'Filtro produto estado{i}{j}'>> beam.Filter(lambda record: (record[2] == produtos[i] and (record[1])==ufs[j]))
        |f'Agregar as Colunas{i}{j}'>> beam.Map(lambda record: (record[2],float(record[4])))
        |f'{i}{j}Construir uma nova tablea derivados'>> beam.CombinePerKey(sum)
        |f'Exibir o re4sultado{i}{j}'>> beam.Map(print)
               
           
    )
    
    p1.run()
```

##Dataflow Derivados do Petróleo

```python
import apache_beam as beam
import os
from apache_beam.options.pipeline_options import PipelineOptions
```

```python
# MODELO GCP

pipeline_options = {
    'project': 'projeto-katia146',
    'runner': 'DataflowRunner',
    'region': 'southamerica-east1',
    'staging_location': 'gs://pipeline_raizen/staging/',
    'temp_location': 'gs://pipeline_raizen/temp/',
    'template_location': 'gs://pipeline_raizen/model/modelo_batch_1'
}

pipeline_options = PipelineOptions.from_dictionary(pipeline_options)

p1 = beam.Pipeline(options=pipeline_options)

derivados = (
    p1
    |'Extrair do CSV'>> beam.io.ReadFromText('gs://pipeline_raizen/dataset_final_derivados.csv',skip_header_lines = 1)
    |'Separador do CSV'>> beam.Map(lambda record: record.split(','))
    |'Filtrar volume'>> beam.Filter(lambda record: float(record[4]) > 0)
    |'Agregar as Colunas'>> beam.Map(lambda record: (record[1],float(record[4])))
    |'Construir uma nova tabela pela soma de volume'>> beam.CombinePerKey(sum)
)

tabela = (
    {'volume total dos derivados':derivados}
    |'Agrupar as pernas' >> beam.CoGroupByKey()
    |'Gravar o resultado'>> beam.io.WriteToText('gs://pipeline_raizen/tratado/agregado',file_name_suffix='.csv') 
)
p1.run()

```
